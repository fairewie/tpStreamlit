# Créez une fonction `scraping_bdm` qui collecte l'ensemble des données des articles présent sur la page d'acceuil et qui renvoi un fichier au format `json`.

```python
def scraping_bdm(art):
    all_articles = []
    for soupape in art:
        try:
            data = {'nom':soupape.find('h3').text,
                    "lien de l'image":soupape.find('img').attrs['src'],
                    'lien de l article':soupape.find('a').attrs['href'],
                    'categorie':soupape.span.text,
                    'date':soupape.find_all('span')[1].text,      
                    }
            all_articles.append(data)
        except:
            pass
    return all_articles
    
    
```

```python
scraping_bdm(article)
```

```python
filename = 'bdm.json'
```

```python
with open(filename, 'w',encoding='utf8') as json_file:
    json.dump(scraping_bdm(article[0]), json_file, 
                ensure_ascii=False,
                        indent=4,  
                        separators=(',',': '))
 
```

```python
article_dict = {}
for art in article:
    id = art.get('id')

    #title
    title = art.find('h3').text

    #image
    try:
        image = art.find('img')['data-lazy-src']
    except:
        image = None
    
    #link
    try:
        link = art.find('a')['href']
    except:
        link = art.parent['href']

    #Catégorie
    try:
        theme = art.find('span', 'favtag').text
    except:
        theme = art.find_previous('h2').text
    
    #date

    date = article[0].find_all('span')[1].text

    article_dict[id] ={
        'title':title,
        'date':date,
        'link':link,
        'image':image,
        'categorie':theme,

    }
```
